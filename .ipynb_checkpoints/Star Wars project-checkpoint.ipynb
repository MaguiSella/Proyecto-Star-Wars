#### Text preprocessing

```python
#https://www.datacamp.com/community/tutorials/text-analytics-beginners-nltk
#https://stackoverflow.com/questions/46786211/counting-the-frequency-of-words-in-a-pandas-data-frame
import nltk
```

As a first step the dialogues are tokenized. 

```python
def tokenize (value):
 return word_tokenize(value)

from nltk.tokenize import word_tokenize
all_episodes['dialogue_tokenized']=all_episodes['dialogue'].apply(tokenize)
```

```python
all_episodes

```

Since the method used generates a list for each line, we merge all the episodes together.

```python
import itertools

merged = list(all_episodes['dialogue_tokenized'])

merged = list(itertools.chain.from_iterable(merged))
merged[:20]
```

As a next step we analyse the word frequency.

```python
from nltk.probability import FreqDist
```

```python
#task 4
fdist1 = FreqDist(merged)
fdist1 = fdist1.most_common(20)
fdist1
```

```python
#PLOT THE MOST COMMON WORDS
from matplotlib.pyplot import figure

pd.DataFrame(fdist1, columns=['lbl','val']).set_index('lbl').plot(kind='bar',figsize=(15, 6))

```

From the plot above it can be easily seen that the most common words are punctuation signs and the to be verb, as well as pronouns. This, however, does not help us understand the scripts. Thus, the scripts must be prepared and cleaned so we can drive insights from this kind of analysis.

```python
import re
import nltk
#from nltk.corpus import stopwords
import nltk as nlp
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

First we process the dilogues performing lemmanization and then without performing it, since it can affect the sentiment analysis profoundly.

```python
#In this step many text cleaning operations are performed together: handling upper and lower case, lemmanizing and tokenization,
#even though we have already performed such an operaration before.
#This is done again since it is simpler to do all the other operations in on the original dialogue rather than on lists.
description_list=[]
for description in all_episodes.dialogue:
    description=re.sub("[^a-zA-Z]", " ", description)
    description=description.lower()
    description=nltk.word_tokenize(description)
    description=[word for word in description if not word in set(stopwords.words("english"))]
    lemma=nlp.WordNetLemmatizer() #we prefer lemmanization over stemming
    description=[lemma.lemmatize(word) for word in description]
    description=" ".join(description)
    description_list.append(description)
```

```python
#create a new column with the processed dialogue
all_episodes["new_script"]=description_list
all_episodes
```

```python
#In this step many text cleaning operations are performed together: handling upper and lower case, lemmanizing and tokenization,
#even though we have already performed such an operaration before.
#This is done again since it is simpler to do all the other operations in on the original dialogue rather than on lists.
description_list=[]
for description in all_episodes.dialogue:
    description=re.sub("[^a-zA-Z]", " ", description)
    description=description.lower()
    description=nltk.word_tokenize(description)
    description=[word for word in description if not word in set(stopwords.words("english"))]
    #lemma=nlp.WordNetLemmatizer() #we prefer lemmanization over stemming
    #description=[lemma.lemmatize(word) for word in description]
    description=" ".join(description)
    description_list.append(description)
```

```python
#create a new column with the processed dialogue
all_episodes["new_script_nolemma"]=description_list
all_episodes
```

```python
#Plot word frequency after cleaning operations

merged = list(all_episodes['new_script'])

fdist2 = FreqDist(merged)
fdist2 = fdist2.most_common(20)

pd.DataFrame(fdist2, columns=['lbl','val']).set_index('lbl').plot(kind='bar',figsize=(15, 6))
plt.show()
```

When performing the same analisis again but over the cleansed dataset we can see that the most common words have changed, with some proper named such as luke and chewie apperaring, but with many common verbs as well, such as come, going, know. Even with this cleansed dataset it is still difficult to drive clear conclusions from it, since the importance of a word is given not ony by how many times it appears in a text, but rather by how many times it appears without being way too common.
```python
#TASK 10
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
pd.set_option('display.max_columns', None)
sentiment = SIA()

```

Polarity scores with lemmatized dialogues

```python
#YODA
yoda1 = ' '.join(list(yoda['new_script']))
yoda_values = list(sentiment.polarity_scores(yoda1).values())
```

```python
#VADER
vader1 = ' '.join(list(vader['new_script']))
vader_values = list(sentiment.polarity_scores(vader1).values())
```

```python
#GENERAL

ae1 = ' '.join(list(all_episodes['new_script']))
ae_values = list(sentiment.polarity_scores(ae1).values())
```

```python
#non processed GENERAL

ae2 = ' '.join(list(all_episodes['dialogue']))
ae2_values = list(sentiment.polarity_scores(ae2).values())
print(ae2_values)
print(sentiment.polarity_scores(ae2))
```

```python
list_of_lists = [yoda_values,vader_values,ae_values,ae2_values]
matrix = np.array(list_of_lists)
matrix = matrix.transpose()
```

```python
matrix
```

```python
df = pd.DataFrame(
{"character" : ['YODA','VADER', 'GENERAL','GENERAL_OR'],
 "neg" : matrix[0,],
"neu" : matrix[1,],
"pos" : matrix[2,]})
```

If the scripts are lemmatized, the following conclusions hold

```python
df
```

```python
fig, ax1 = plt.subplots(figsize=(5, 5))
tidy = df.melt(id_vars='character').rename(columns=str.title)
sns.barplot(x='Character', y='Value', hue='Variable', data=tidy, ax=ax1,palette={'pos':'g', 'neg':'k', 'neu':'#C0C0C0'})
sns.despine(fig)
```

As can be seen on the plot above, the polarity scores for Yoda and Vader are not quite different, with Yoda actually having a negative polarity score slightly greater than Vader, and having the greatest when compared with the thrilogy in general. 
This is counterintuitive, since one would expect the good character to have a smaller negative polarity score than the bad one.
In general, the movie has a net positive polarity scores, as do the two characters analyzed. 
This analysis may be biased by the library used to perform sentiment analysis, which is more suited to social media analysis.  

However, when using the Vader sentiment analysis algorithm, since it gives different values in terms of the endings of the words, lemmatizing is not recommended. In this case, the positive and negative polarity scores increase substantially, and the neuter one shrinks in all the cases.
